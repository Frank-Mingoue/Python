```python
tables = df.TABLE_NAME.str.strip().unique()
tables
```

```python
stuct_def = pd.DataFrame({"table_name" : [], "table_struct" : []})
for table in tables :
    df_table = df[df.TABLE_NAME.str.strip() == table]
    strutfields = ""
    for i, v in df_table["FIELD"].items():
        strutfields = strutfields + v + '\n \t'

    StructType = f"#------------{table} \n" + f"""elif filename == "{table}" : \n \t""" + "schema = StructType( [\n" +f"\t{strutfields}"  + """]) \n\n \treturn schema"""

    stuct_def = stuct_def.append({"table_name": table, "table_struct":StructType}, ignore_index=True)
    #print(strutfields)
```

```python
stuct_def.to_excel("strutFile.xlsx")
stuct_def.to_csv("strutFile.csv")
```
