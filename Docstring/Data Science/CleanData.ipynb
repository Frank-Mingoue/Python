Traiter les valeurs manquantes

```python
df = pd.read_csv("data.csv")
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df[df.tax.isnull()]
```

```python
df.isnull().any(axis="columns")
```

```python
df[df.isnull().any(axis="columns")]
```

```python
df.tax.fillna(0, inplace=True)
```

```python
df
```

```python
df.dropna(subset=['gender', 'country'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.price_paid = df.price_paid.apply(lambda x : x.replace("$", ""))
df.price_paid = df.price_paid.astype(float)
df
```

```python
df["produc_price"] = df["price_paid"]  * (1 - df["tax"] / 100) 
```

```python

df["produc_price"] = df["produc_price"].rename("product_price", inplace=True)
df
```

```python
df.query("country =='Canada'")
```

```python
df.country.unique()
```

```python
country_code = {'France': 'FR', 'Canada': 'CA', 'United States': 'US', 'Morocco': "MO"  }

df["coutry_code"] =  df.country.apply(lambda x : country_code[x])
```

```python
df
```

```python
df.rename(columns={"produc_price": "product_price", "coutry_code": "country_code"}, inplace=True)
```
